```python
# Compare average duration of calls per each plan per each distinct month. Plot a bar plat to visualize it.
avg_duration = merged_data.groupby(['plan', 'month'])['total_duration'].mean().reset_index()

pivot_duration = avg_duration.pivot(
    index='month',
    columns='plan',
    values='total_duration'
)

# Plot bar chart
pivot_duration.plot(
    kind='bar',
    figsize=(10, 5)
)

plt.title('Average Call Duration per Month by Plan')
plt.xlabel('Month')
plt.ylabel('Average Total Duration')
plt.xticks(rotation=0)
plt.legend(title='Plan')

plt.show()
```

[Test the hypothesis that the average revenue from users of the Ultimate and Surf calling plans differs.]

[Formulate the null and the alternative hypotheses, choose the statistical test, decide on the alpha value.]

```python

# Test the hypotheses
# Revenue by plan

surf_revenue = merged_data[merged_data['plan'] == 'surf']['revenue'].dropna()
ultimate_revenue = merged_data[merged_data['plan'] == 'ultimate']['revenue'].dropna()

# Check sample sizes
print(f"Surf sample size: {len(surf_revenue)}")
print(f"Ultimate sample size: {len(ultimate_revenue)}")

# Descriptive statistics
print(f"\nSurf - Mean: ${surf_revenue.mean():.2f}, Std: ${surf_revenue.std():.2f}")
print(f"Ultimate - Mean: ${ultimate_revenue.mean():.2f}, Std: ${ultimate_revenue.std():.2f}")

# Perform independent samples t-test (Welch's t-test)
results = st.ttest_ind(
    surf_revenue,
    ultimate_revenue,
    equal_var=False  # Welch's t-test (doesn't assume equal variances)
)

# Alpha level
alpha = 0.05

# Display results
print(f"\n--- T-Test Results ---")
print(f"t-statistic: {results.statistic:.4f}")
print(f"p-value: {results.pvalue:.6f}")
print(f"Alpha level: {alpha}")

if results.pvalue < alpha:
    print(f"\n✓ Reject the null hypothesis (p < {alpha})")
    print("There is a statistically significant difference between plans. Acept H1 Alternative-Hypothesis")
else:
    print(f"\n✗ Fail to reject the null hypothesis (p >= {alpha})")
    print("No statistically significant difference found. Acept HO null-Hypothesis")

# Calculate effect size (Cohen's d)
pooled_std = np.sqrt(((len(surf_revenue)-1)*surf_revenue.std()**2 + 
                       (len(ultimate_revenue)-1)*ultimate_revenue.std()**2) / 
                      (len(surf_revenue) + len(ultimate_revenue) - 2))
cohens_d = (surf_revenue.mean() - ultimate_revenue.mean()) / pooled_std
print(f"\nEffect size (Cohen's d): {cohens_d:.4f}")

```
